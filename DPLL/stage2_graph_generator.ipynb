```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = "data.csv"
```

```python
df = pd.read_csv(filename, header=0, delimiter=',')
```

```python
df3 = df.loc[df['K'] == 3]
df4 = df.loc[df['K'] == 4]
df5 = df.loc[df['K'] == 5]
```

```python
df3
```

```python
plt.plot( 'R', 'Stats', data=df3, marker='', color='red', linewidth=2, label="K=3")
plt.xlabel("R value")
plt.ylabel("Probability of satisfiability")
plt.show()
```

```python
plt.plot( 'R', 'Stats', data=df4, marker='', color='blue', linewidth=2, label="K=4")
plt.xlabel("R value")
plt.ylabel("Probability of satisfiability")
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
plt.plot( 'R', 'Stats', data=df5, marker='', color='green', linewidth=2, label="K=5")
plt.xlabel("R value")
plt.ylabel("Probability of satisfiability")
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
plt.plot( 'R', 'Stats', data=df3, marker='', color='red', linewidth=2, label="K=3")
plt.plot( 'R', 'Stats', data=df4, marker='', color='blue', linewidth=2, label="K=4")
plt.plot( 'R', 'Stats', data=df5, marker='', color='green', linewidth=2, label="K=5")
plt.xlabel("R value")
plt.ylabel("Probability of satisfiability")
plt.legend()
plt.show()
```
